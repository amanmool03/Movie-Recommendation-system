But this is in the form of string and we cant work pass it as an object to extract the keys we want therefore using a builtin lib called ast we convert the string into list 

```python
import ast
# ast.literal_eval()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

Aba the op of genres from  [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] is like the following

```python
movies.head()
```

Likewise repeat the helper function for all the remaining columns that we are trying to merge into a new column called TAGS

```python
movies['keywords']=movies['keywords'].apply(convert)
```

But we need to extract top 3 cast from the dict so change the helper function accordingly

```python
def convertCast(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convertCast)
```

now to extract the directors name 

```python
print(movies['crew'][0])
```

```python
def convertDirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=="Director"):
            L.append(i['name'])
        else:
            pass
    return L
```

```python
movies['crew']=movies['crew'].apply(convertDirector)
```

```python
print(movies['crew'][0])
```

since overview is a string convvert it to list so that we can concatenate it with the other columns

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

now remove spaces in between a single entity that is because a machine can not seperate Sam X and Sam Y on its own , to make it easier for the computer to understand we merge them into a single entity ie SamX and SamY

```python
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

now generating the TAGS column by merging the following columns

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

now converting the tags list into a string and also making them lowercase

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x).lower())
```

# Stemming   [loved,loves,loving]=[love,love,love]

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

stemmer helper function

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

'in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission....'    ➡️  TO     'in the 22nd century, a parapleg marin is dispatch to the moon pandora on a uniqu mission.....'

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

Repeat vectorization again

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

Euclean Distance vaneko duita vector ko end point to endpoint ko distance
but hamile tyo na nikalne instead
cosine distance or angle between those two vectors chai nikalne

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

Harek movie bata harek movie sanga ko distance ie. arrays of arrays

```python
cosine_similarity(vectors).shape
```

Enumerate to add indexes to the list

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

now create a helper function that will genereate the id of the movie user entered obtain corresponding array sort it and display the top 5's

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```

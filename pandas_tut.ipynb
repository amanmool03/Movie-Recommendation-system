# The main distinction between loc and iloc is: loc is label-based, which means that you have to specify rows and columns based on their row and column labels. iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

```python
##Sorting
df.sort_values('Attack',ascending=False)
```

```python
#Pokemon containing Mega in their names
df.loc[df['Name'].str.contains('Mega')]
```

```python
#Pokemon without Megas in their names
df.loc[~df['Name'].str.contains('Mega')]
```

#We can also enter Regular Expresions which is very powerful 

```python
import re
df.loc[df['Type 1'].str.contains('Fairy|Ice',regex=True)]
```

```python
#Printing all the Pokemons Starting With PI using REGEX filtering

df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]

#  ^ MEANS start of
#  pi[a-z]* means pi folowing 0 or more a-z alphabets
```

# Conditional Changes changing Values of columns of Datas

#Conditional Changes  Changing Bugs to Buggers

```python

df.loc[df['Type 1']=="Bug",'Type 1']="Buggers"
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]

```

```python
#NAN replace vayena
df.loc[df['Type 2']=="NaN",'Type 2']="X"
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

# GroupBy method

```python
df.groupby(['Type 1']).mean().sort_values('Total_2',ascending=False)
```

```python
Stats=df.groupby(['Type 1']).mean().sort_values('Total_2',ascending=False)
Stats.to_csv("StatsOnPokemons.csv")
```
